4)write a python program to print out the names which exists in the first name list but not exist in the second set of name list as same answer in the following output ?   Output is: name_list_1= “Nisam”, “Ash”, “Jason”, “Pranav” name_list_2= “Pranav”, “Jason” Answer: The names which exists in the first name list but not exist in the second set of name list : {‘Nisam’, ‘Ash’}

```python
name_list_1 = ["Nisam", "Ash", "Jason", "Pranav"]
name_list_2 = ["Pranav", "Jason"]

print('name_list_1 = "Nisam", "Ash", "Jason", "Pranav" ') 
print('name_list_2 = "Jason", "Pranav" ') 

x = set(name_list_1) - set(name_list_2)
print(f"The names which exists in the first name list but not exist in the second set of name list : {x}")
```

6)Write a python program to allow user to input their weight and height in order to calculate the BMI of user. [formula of BMI = weight (kg) / height (m)]?

Output is:

Please input your weight in kg: 50 Please input your height in m: 1.56

Your BMI is 20.55

```python
Weight = float(input("please input weight in kg : ")) 
Height = float(input("please input height in m : ")) 

BMI = Weight/(Height**2) 

print(f"Your BMI is {BMI}")
```
