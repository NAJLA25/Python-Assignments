1. Write multiple programs which illustrates iteration through various data structures which you learned till now?

```python
# if else elif

a = 10
if a < 10:
  print("this is a valid statement")
elif a == 10:
  print("this is a valid statement from elif")
else:
 print("this is a default statement")
```

```python
#nested if

i = 10
n = "najla"
if i == 10:
  if type(n) == int:
    print("this is my name")
  else:
    print("this is a default condition")
else:
  print("default statement from outer loop")
```

```python
#input

num = int(input("enter a number : "))
if num >= 1:
  print("this is a positive number") 
elif num < 0:
  print("this is a negative number") 
else:
  print("this is zero")
```

```python
#for loop

listt = [22, 33, 44, 55, 66, 77]
for i in listt:
  print(i)
else:
  print("no more numbers")
```

```python
#if inside for loop

sum = 0
for i in listt:
  if type(i) == int:
    sum = sum + i
sum
```

```python
#break

for x in listt:
  if x == 55:
    break
  print(x) 
else:
  print("no more numbers")
```

```python
#continue

for x in listt:
  if x == 55:
    continue
  print(x)
else:
  print("no more numbers")
```

```python
#range

for i in range(0, 50, 5):
  print(i)
```

```python
#len

for i in range(len(listt)):
  print(i)
```

```python
#append

string = "hello world"
l = []
l1 = []
for i in range(len(string)):
  if string[i] == 'l' :
    l.append(i) 
    l1.append(string[i])
print(l, l1, sep = '\n')
```

2. Receive an input text from the user and print all letters except ‘e’ and ‘s’ by using for loop?

```python

text = str(input("enter a text : "))
for i in text :
  if (i == "e"):
    continue
  elif (i == "s") :
    continue
  else :
    print(i)
```

3. Return sum of first n numbers with the help of for loop and range function?

```python
number = int(input("Enter the Number: "))
sum = 0
for n in range(1, number+1):
    sum = sum + n
print(sum)
```
